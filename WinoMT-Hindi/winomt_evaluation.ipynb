```python
def read_data(filename):
    with open(filename, 'r') as f:
        data = [line.split('\t') for line in f.read().splitlines()]
    return data
```

```python
def read_data_list(filename):
    data = []
    with open(filename, 'r') as f:
        a = [line.split('\t') for line in f.read().splitlines()]
        for i in range(0, len(a)):
            data.append(a[i][0])
    return data
```

```python
import math
import nltk
```

```python
def evaluatee(data, golds):
    labels = []
    neul = 0
    m = 0
    f = 0
    for i, s in enumerate(data):
        tokens = nltk.word_tokenize(s.lower())
        
        if 'he' in tokens and 'she' in tokens:
                print(s)
                label = str(input("Enter whether - male or female or neutral : ")) #happens rarely
                if label == 'neutral':
                    neul+=1
                    if golds[i] == 'male':
                        m+=1
                    else :
                        f+=1

                labels.append(label)

        elif 'she' in tokens:
            labels.append('female')

        elif 'he' in tokens:
            labels.append('male')

        elif 'his' in tokens or 'him' in tokens:
            if 'her' in tokens:
                print("Mixed : ",i+1 ,s)
                label = str(input("Enter whether - male or female : ")) #happens rarely
                labels.append(label)

            else:
                labels.append('male')
                continue

        elif 'her' in tokens:
            labels.append('female')
            continue

        else:
            labels.append('neutral')
            neul+=1
            if golds[i] == 'male':
                m+=1
            else :
                f+=1

    print()
    print("No of neutral sentences : ", neul)
    print("percentage neutral",neul/len(data)*100 )
    print()
    print("Out of neutral sentences, no of actually male : ", m)
    print("percentahe neutral actually male", m/neul*100)
    print()
    print("Out of neutral sentences, no of actually female : ", f)
    print("percentahe neutral actually female", f/neul*100)

    return labels
```

## Google Evaluation

```python
#google evaluation

# opening the file in read mode
my_file_anti = open("stereo_google/antistereohindi_google.txt", "r")
  
# reading the file
data = my_file_anti.read()
  
# replacing end of line('/n') with ' ' and
# splitting the text it further when '.' is seen.
data_into_list_anti = data.split("\n")
  
# printing the data
print(data_into_list_anti)
my_file_anti.close()
```

```python
#google evaluation

# opening the file in read mode
my_file_pro = open("stereo_google/prostereohindi_google.txt", "r")
  
# reading the file
data = my_file_pro.read()
  
# replacing end of line('/n') with ' ' and
# splitting the text it further when '.' is seen.
data_into_list_pro = data.split("\n")
  
# printing the data
print(data_into_list_pro)
my_file_pro.close()

anti_preds = evaluatee(data_into_list_anti, anti_golds)
```

```python
((18+19)/704)*100
```

```python
pro_preds = evaluatee(data_into_list_pro, pro_golds)
```

```python
len(anti_preds)
```

```python
len(pro_preds)
```

```python
from sklearn.metrics import accuracy_score

ground_truth_labels = anti_golds + pro_golds
predicted_labels = anti_preds + pro_preds

# calculate accuracy
accuracy = accuracy_score(ground_truth_labels, predicted_labels)
print("acc percentage")
print(accuracy*100)
```

```python
from sklearn.metrics import precision_recall_fscore_support


precision_male, recall_male, f1_score_male, _ = precision_recall_fscore_support(ground_truth_labels, predicted_labels, average=None, labels=['male'])
print("f1 for male : ", f1_score_male[0])

precision_female, recall_female, f1_score_female, _ = precision_recall_fscore_support(ground_truth_labels, predicted_labels, average=None, labels=['female'])
print("f1 for female : ", f1_score_female[0])

print("delta G measure : ", f1_score_male[0]  - f1_score_female[0])
```

```python
def calculate_macro_averaged_f1_score(gold_labels, predicted_labels):
    precision_male, recall_male, f1_score_male, _ = precision_recall_fscore_support(gold_labels, predicted_labels, average=None, labels=['male'])
    precision_female, recall_female, f1_score_female, _ = precision_recall_fscore_support(gold_labels, predicted_labels, average=None, labels=['female'])
    macro_f1 = (f1_score_male[0] + f1_score_female[0])/2*100
    return macro_f1

# Example usage
anti_stereotypical_gold_labels = anti_golds
anti_stereotypical_predicted_labels = anti_preds

pro_stereotypical_gold_labels = pro_golds
pro_stereotypical_predicted_labels = pro_preds

anti_stereotypical_macro_f1 = calculate_macro_averaged_f1_score(anti_stereotypical_gold_labels, anti_stereotypical_predicted_labels)
pro_stereotypical_macro_f1 = calculate_macro_averaged_f1_score(pro_stereotypical_gold_labels, pro_stereotypical_predicted_labels)

print(f"Macro-Averaged F1 Score for Anti-Stereotypical Sentences: {anti_stereotypical_macro_f1:.2f}")
print(f"Macro-Averaged F1 Score for Pro-Stereotypical Sentences: {pro_stereotypical_macro_f1:.2f}")

print("dela S : ", pro_stereotypical_macro_f1 - anti_stereotypical_macro_f1)
```
